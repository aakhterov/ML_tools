```python
import pandas as pd
import numpy as np
```

```python
# Load data and look at the data
```

```python
df=pd.read_csv("weatherAUS.csv")
df.info()
```

```python
df.head()
```

```python
# Checking fraction of NaN value across columns
```

```python
for col in df.columns:
    print(f"{col}: {(round(df[col].isnull().mean()*100))}% NaN")
```

```python
# For followed columns replacing Nan value by mode() value
```

```python
cols_NaN_to_mode = ["MinTemp", "MinTemp", "Rainfall", "WindGustDir", "WindGustSpeed", 
                    "WindDir9am", "WindDir3pm", "WindSpeed9am", "WindSpeed3pm", 
                    "Humidity9am", "Humidity3pm", "Pressure9am", "Pressure3pm", 
                    "Temp9am", "Temp3pm"]
for col in cols_NaN_to_mode:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# For Cloud9am and Cloud3pm columns replacing Nan value by mean() value
```

```python
df["Cloud9am"].fillna(round(df["Cloud9am"].mean()), inplace=True)
df["Cloud3pm"].fillna(round(df["Cloud3pm"].mean()), inplace=True)
```

```python
# Remove row with NaN value in columns RainTomorrow and RainToday and remove columns Evaporation and Sunshine 
# because there are many NaN values - 43% and 48% respectively
```

```python
df.dropna(axis=0, subset=["RainTomorrow", "RainToday"], inplace=True)
df.drop(columns=["Evaporation", "Sunshine"], inplace=True)
```

```python
# Looking at the data again 
```

```python
df.info()
```

```python
# We got a good dataset without NaN values, but still remain some details
```

```python
# Change type of Date, RainToday and RainTomorrow columns
```

```python
df["Date"] = df["Date"].astype("datetime64")
df["RainToday"] = df["RainToday"].astype("category")
df["RainTomorrow"] = df["RainTomorrow"].astype("category")
```

```python
# Replace value Yes and No to 1 and 0 respectively in columns RainToday and RainTomorrow
```

```python
class_to_num = {"RainToday": {"Yes": 1, "No": 0}, "RainTomorrow": {"Yes": 1, "No": 0}}
df.replace(class_to_num, inplace=True)
```

```python
# Reset index
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# Applying function for creating new columns
```

```python
def diff_temperature(row):
    return row.MaxTemp - row.MinTemp
```

```python
df["DiffTemp"] = df.apply(diff_temperature, axis=1)
```

```python
# Creating CrossTab and adding a new columns
```

```python
target_to_location_df = pd.crosstab(df.Location, df.Target)
target_to_location_df.head()
```

```python
target_to_location_df["FracOfRainDay"] = target_to_location_df[1]/(target_to_location_df[0] + target_to_location_df[1])
```

```python
target_to_location_df.sort_values("FracOfRainDay")
```
